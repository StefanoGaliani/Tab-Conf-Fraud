```python
import pandas as pd
import numpy as np
from centroid import centroid, D  # Custom modules
from pathlib import Path
import os
import glob
from concurrent.futures import ThreadPoolExecutor
```

```python
def test(filename, t=2, alpha=0.5, delta=0.8, gamma=0.01):
    
    df = pd.read_excel(filename)
    print(f"Processing {os.path.basename(filename)}")

    acc_centroid = acc_amaxmin = acc_penalized = acc_penalized_w = acc_wborda = 0
    TP_centroid = FN_centroid = 0
    TP_amaxmin = FN_amaxmin = 0
    TP_penalized = FN_penalized = 0
    TP_penalized_w = FN_penalized_w = 0
    TP_wborda = FN_wborda = 0

    type_div = t  # 1: KL-divergence, 2: L2-squared

    for i in range(len(df)):
        actual = int(df.iloc[i]['Actual'])
        distributions = []
        weights = []

        for m in range(1, 33):
            p = df.iloc[i]['Prob_' + str(m)]
            c = df.iloc[i]['Conf_' + str(m)]
            distributions.append(np.array([1 - p, p]))
            weights.append(c)

        distributions = np.array(distributions)
        weights = np.array(weights)

        # --- Centroid computation ---
        opt_p, _ = centroid(distributions, weights, t=type_div)

        # --- Distance filtering ---
        distances = [D(opt_p, dist, t=type_div) for dist in distributions]
        distances = np.array(distances)
        quantile = np.quantile(distances, delta)
        mask = distances <= quantile
        filtered_distribs = distributions[mask]
        filtered_weights = weights[mask]

        # --- Centroid classification ---
        classified = np.argmax(opt_p)
        if actual == classified:
            acc_centroid += 1
            if actual == 1:
                TP_centroid += 1
        elif actual == 1:
            FN_centroid += 1

        # --- Alpha-maxmin classification ---
        min_dist = filtered_distribs.min(axis=0)
        max_dist = filtered_distribs.max(axis=0)
        alpha_mix = alpha * min_dist + (1 - alpha) * max_dist
        classified = np.argmax(alpha_mix)
        if actual == classified:
            acc_amaxmin += 1
            if actual == 1:
                TP_amaxmin += 1
        elif actual == 1:
            FN_amaxmin += 1

        # --- Penalized mean softmax (simple) ---
        mean_softmax = filtered_distribs.mean(axis=0)
        pen_mean_softmax = mean_softmax * (1 - gamma * alpha_mix)
        classified = np.argmax(pen_mean_softmax)
        if actual == classified:
            acc_penalized += 1
            if actual == 1:
                TP_penalized += 1
        elif actual == 1:
            FN_penalized += 1

        # --- Penalized mean softmax (weighted) ---
        w = (filtered_weights / filtered_weights.sum()).reshape(len(filtered_weights), 1)
        mean_softmax = w.T @ filtered_distribs
        pen_mean_softmax = mean_softmax * (1 - gamma * alpha_mix)
        classified = np.argmax(pen_mean_softmax)
        if actual == classified:
            acc_penalized_w += 1
            if actual == 1:
                TP_penalized_w += 1
        elif actual == 1:
            FN_penalized_w += 1

        # --- Weighted Borda classification ---
        ranks = np.array([
            np.argmin(filtered_distribs, axis=1),
            np.argmax(filtered_distribs, axis=1)
        ]).T
        wborda = np.array([
            ranks[:, 0].dot(filtered_weights),
            ranks[:, 1].dot(filtered_weights)
        ])
        classified = np.argmax(wborda)
        if actual == classified:
            acc_wborda += 1
            if actual == 1:
                TP_wborda += 1
        elif actual == 1:
            FN_wborda += 1

    result = {
        "filename": os.path.basename(filename),
        "centroid": {
            "accuracy": round(acc_centroid / len(df) * 100, 1),
            "recall": round(TP_centroid / (TP_centroid + FN_centroid + 1e-8) * 100, 1)
        },
        "Alpha-maxmin classification": {
            "accuracy": round(acc_amaxmin / len(df) * 100, 1),
            "recall": round(TP_amaxmin / (TP_amaxmin + FN_amaxmin + 1e-8) * 100, 1)
        },
        "Penalized mean softmax (simple mean)": {
            "accuracy": round(acc_penalized / len(df) * 100, 1),
            "recall": round(TP_penalized / (TP_penalized + FN_penalized + 1e-8) * 100, 1)
        },
        "Penalized mean softmax (weighted mean)": {
            "accuracy": round(acc_penalized_w / len(df) * 100, 1),
            "recall": round(TP_penalized_w / (TP_penalized_w + FN_penalized_w + 1e-8) * 100, 1)
        },
        "Weighted Borda classification": {
            "accuracy": round(acc_wborda / len(df) * 100, 1),
            "recall": round(TP_wborda / (TP_wborda + FN_wborda + 1e-8) * 100, 1)
        }
    }

    return result

    # # --- Results summary ---
    # print('*** Centroid classification ***')
    # print('Accuracy:', round(acc_centroid / len(df) * 100, 1))
    # print('Recall  :', round(TP_centroid / (TP_centroid + FN_centroid) * 100, 1))
    # print()
    # print('*** Alpha-maxmin classification ***')
    # print('Accuracy:', round(acc_amaxmin / len(df) * 100, 1))
    # print('Recall  :', round(TP_amaxmin / (TP_amaxmin + FN_amaxmin) * 100, 1))
    # print()
    # print('*** Penalized mean softmax (simple mean) ***')
    # print('Accuracy:', round(acc_penalized / len(df) * 100, 1))
    # print('Recall  :', round(TP_penalized / (TP_penalized + FN_penalized) * 100, 1))
    # print()
    # print('*** Penalized mean softmax (weighted mean) ***')
    # print('Accuracy:', round(acc_penalized_w / len(df) * 100, 1))
    # print('Recall  :', round(TP_penalized_w / (TP_penalized_w + FN_penalized_w) * 100, 1))
    # print()
    # print('*** Weighted Borda classification ***')
    # print('Accuracy:', round(acc_wborda / len(df) * 100, 1))
    # print('Recall  :', round(TP_wborda / (TP_wborda + FN_wborda) * 100, 1))

    
```

```python
def test_wrapper(filename):
    try:
        return test(filename)
    except Exception as e:
        return {"filename": os.path.basename(filename), "error": str(e)}

file_source = "/Users/stefanogaliani/Desktop/Tabular_Fraud/TheBigUnknown/Excel Model Output"
excel_files = list(Path(file_source).glob("*.xlsx"))

with ThreadPoolExecutor() as executor:
    results = list(executor.map(test_wrapper, excel_files))

# Display
for i, r in enumerate(results):
    print("=" * 40)
    print(f"📁 File: {r.get('filename', 'N/A')}")
    print("-" * 40)

    if "error" in r:
        print(f"❌ Error: {r['error']}")
    else:
        for method, metrics in r.items():
            if method == "filename":
                continue
            print(f"🔸 {method.capitalize()} → Acc: {metrics['accuracy']}% | Recall: {metrics['recall']}%")

    print("=" * 40 + "\n")

directory_path = Path(file_source)
output_path = directory_path / "results_summary.txt"


with open(output_path, "w", encoding="utf-8") as f:
    for i, r in enumerate(results):
        f.write("=" * 40 + "\n")
        f.write(f"📁 File: {r.get('filename', 'N/A')}\n")
        f.write("-" * 40 + "\n")

        if "error" in r:
            f.write(f"❌ Error: {r['error']}\n")
        else:
            for method, metrics in r.items():
                if method == "filename":
                    continue
                f.write(f"🔸 {method.capitalize()} → Acc: {metrics['accuracy']}% | Recall: {metrics['recall']}%\n")

        f.write("=" * 40 + "\n\n")

print(f"✅ Results saved to: {output_path}")

```
